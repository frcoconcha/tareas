## Problema 02
Nuevamente se utiliza **for** y **range** para imprimir los números, pero esta vez se le pone una condición con **if** para que solo imprima los números divisibles de 6. Para esto, se le agrega el **%**, el cual nos arrojará cada valor de x que cumpla la condición. Finalmente, se imprime con **print**.

```python
for x in range(-100, 201):
  if x % 6 == 0:
    print(x)
```

## Problema 04

En primer lugar, se utiliza **int** e **input** para que el usuario pueda interactuar y agregar su edad, la cual debe ser un número entero. En segundo lugar, se utiliza **input** para que el usuario indique si le gusta la música urbana. Se le indica al programa que si responde "sí" es equivalente a "gusta_musica_urbana" y si responde "no" es igual a "no_gusta_musica_urbana". Finalmente, se utiliza **if** y **elif** para que el programa indique distintas frases según el resultado. Se agrega un **else** para que si ninguna condición se cumple o responde algo que no corresponde arroje un mensaje de que algo pasó.

```python
edad = int(input("Ingresa tu edad"))
musica_urbana = input("¿Te gusta la música urbana? (responde sí o no)")

gusta_musica_urbana = musica_urbana == "sí"
no_gusta_musica_urbana = musica_urbana == "no"

if(edad>=18 and gusta_musica_urbana):
  print("Eres mayor de edad y te gusta la música urbana")
elif(edad<18 and gusta_musica_urbana):
  print("Eres menor de edad y te gusta la música urbana")
elif(edad>=18 and no_gusta_musica_urbana):
  print("Eres mayor de edad y no te gusta la música urbana")
elif(edad<18 and no_gusta_musica_urbana):
  print("Eres menor de edad y no te gusta la música urbana")
else:
  print("Algo pasó")
```
